```python
def bb_from_rlmask(rl_list):
    #list of run (start lenght format)
    x=[]
    y=[]
    #for each row construct a row bounding box
    for i in range(len(rl_list)):
        start = float(rl_list[i].split(" ")[0])
        lenght = float(rl_list[i].split(" ")[1])
        #Rows of run extrema
        x.append(start / size)
        x.append((start+lenght) / size)
        y.append(start % size)
        y.append((start + lenght) % size)
    xmin=int(round(min(x)))
    xmax=int(round(max(x)))
    ymin=int(round(min(y)))
    ymax=int(round(max(y)))
    print(xmin,xmax,ymin,ymax)
    return (xmin,xmax,ymin,ymax) #left, right, top, bottom
```

```python
def get_overlap_mask(bs):
    boxes = bs
    #Order by x then by y coordinate of upper left corner
    boxes = sorted(bs,key=lambda box: (box[0],box[2]))
    nboxes = len(boxes)
    omask = np.zeros((size, size))

    for i in range(0,nboxes):
        for j in range(i+1,nboxes):
            #If there is overlap
            xmin=max(boxes[i][0],boxes[j][0])
            xmax=min(boxes[i][2],boxes[j][2])
            ymin=max(boxes[i][1],boxes[j][1])
            ymax=min(boxes[i][3],boxes[j][3])
            if xmin <= xmax and ymin <= ymax:
                omask[ymin:ymax+1,xmin:xmax+1] = 1
    return omask
```

```python
def rlmask_from_boxes(boxes,scores):
    rlm=[]
    #Fetch unnormalized coordinates
    #(xmin,ymin,xmax,ymax)
    #boxes=[(int(bb[0]*size),int(bb[1]*size),int(bb[2]*size),int(bb[3]*size)) for bb in boxes]
    scores=[max(ss) for ss in scores]
    #Sort by scores
    sbs = zip(scores,boxes)
    sbs.sort(key=lambda sb: -sb[0])
    boxes = list(map(lambda sb: sb[1],sbs))
    print(sbs)
    print(boxes)
    omask = get_overlap_mask(boxes)
    nboxes = len(boxes)
    b = 0
    #Filter out boxes with complete overlap
    while b < nboxes:
        box = boxes[b]
        smask = np.zeros((size, size))
        smask[box[1]:box[3]+1,box[0]:box[2]+1] = 1
        smask = np.subtract(smask,omask)
        smask[smask!=1] = 0
        #Ordered by rows
        rows,cols = np.nonzero(smask) 
        if len(rows)==0:
            #Complete overlap: delete box and update overlap mask
            #print("Deleted" + str(box))
	    for bb in range(b,nboxes-1):
                boxes[bb] = boxes[bb+1]
            nboxes = nboxes - 1
            print(boxes[0:nboxes])
            omask = get_overlap_mask(boxes[0:nboxes])
            b = b-1
        b = b+1
    print(boxes[0:nboxes])
    #Get runs removing overlapping areas
    b = 0
    while b < nboxes:
	box = boxes[b]
        smask = np.zeros((size, size))
        smask[box[1]:box[3]+1,box[0]:box[2]+1] = 1
        smask = np.subtract(smask,omask)
        smask[smask!=1] = 0
        #Ordered by rows
        rows,cols = np.nonzero(smask) 
        #if len(rows)!=0:
        cr = zip(cols,rows)
        #Ordered by columns
        cr.sort(key=lambda x: (x[0],x[1]))
        cols=list(map(lambda x: x[0],cr))
        rows=list(map(lambda x: x[1],cr))
        ship = ""
        start = size*cols[0]+rows[0]
        i = 0
        while i<len(rows):
            lenght=0
            flag = True
            while flag:
                lenght = lenght+1
                i = i+1
                #Same column, sequential rows or sequential columns and peripheal rows => same run
                flag = i<len(rows) and ((cols[i] == cols[i-1] and rows[i] == rows[i-1]+1) or
                                   (cols[i] == cols[i-1]+1 and rows[i-1]==size-1 and rows[i]==0))
            #Close run
            ship=ship+" "+str(start)+" "+str(lenght)  
	    #New run
            if i<len(rows):
                start = size*cols[i]+rows[i]
        ship=ship[1:]
        rlm.append(ship)
        #Next box
        b = b+1	
    return rlm
```

```python
boxes = [(2,2,6,5),(5,1,7,2),(7,2,8,3),(7,2,7,2),(4,5,5,6),(5,2,6,2)]
scores = [[1],[2],[3],[4],[5],[6]]
rlmask_from_boxes(boxes,scores)
```

```python
def prova(id):
    import matplotlib.patches as patches
    import matplotlib.pyplot as plt
    from PIL import Image
    size=768
    filename,rl_string=contents1[id].split(",")
    print(filename)
    rl_list=re.findall('(\d+\ \d+)+',rl_string)
    
    bb_string=contents2[id].split(",")[1]
    bb=re.findall('\d\.\d+',bb_string)
    #print(rl_string)
    #print(boxes_to_rlms([(float(bb[0]),float(bb[1]),float(bb[2]),float(bb[3]))]))
    bb=(int(float(bb[0])*768),int(float(bb[1])*768),int(float(bb[2])*768),int(float(bb[3])*768))
    print(bb)
    rect1=patches.Rectangle((bb[0],bb[3]),bb[1]-bb[0],bb[2]-bb[3],linewidth=1,edgecolor='r',facecolor='none')
    bb=bb_from_rlmask(rl_list)
    print(bb)
    rect2=patches.Rectangle((bb[0],bb[3]),bb[1]-bb[0],bb[2]-bb[3],linewidth=1,edgecolor='b',facecolor='none')

    input_path = "./Challenge/test/"
    image = Image.open(input_path+filename)
    fig,ax = plt.subplots(1)
    ax.add_patch(rect1)
    ax.add_patch(rect2)
    ax.imshow(image)
    plt.show()
```
